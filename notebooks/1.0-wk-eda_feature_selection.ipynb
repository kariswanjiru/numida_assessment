## import libraries and load data

## Feature Engineering

```python
# Replace missing values with "none" in column: 'dismissal_description'
data = data.fillna({'dismissal_description': "none"})
```

```python
# Replace all instances of "none" with "1" in column: 'dismissal_description'
#data.loc[data['dismissal_description'].str.lower() == "none".lower(), 'dismissal_description'] = "1"
```

```python
# Replace missing values with 0 in column: 'employee_count'
data = data.fillna({'employee_count': 0})
```

```python
null_rows = data[data['total_owing_at_issue'].isnull()]
null_rows
```

```python
null_rows = data[data['loan_number'].isnull()]
null_rows
```

```python
data['payment_status'] = data['principal']-data['amount_paid']
data
```
