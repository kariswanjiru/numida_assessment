## Feature Engineering and Selection

### Missing values

```python
# Replace missing values with "none" in column: 'dismissal_description' if the appoval status is apporved.
def missing_dismissal(df):
    if df['approval_status']=='Approved':
        return "none:approved loan."
    if df['approval_status']=='Declined':
        return  "declined"
    if df['approval_status']=='Cancelled':
        return "application cancelled."
    if df['approval_status']=='Expired':
        return "application expired"

loan['dismissal_description'] = loan['dismissal_description'].fillna(loan.apply(missing_dismissal, axis=1))
test['dismissal_description'] = test['dismissal_description'].fillna(loan.apply(missing_dismissal, axis=1))
loan.head(5).style
```

The features total_owing_at_issue,total_recovered_on_time,total_recovered_15_dpd,cash_yield_15_dpd,loan_number are null when loan is not approved, hence replace the null values as 0 instaed of dropping null values.

```python
# Replace missing values with 0 in column: 'total_owing_at_issue'
loan = loan.fillna({'total_owing_at_issue': 0})
test = test.fillna({'total_owing_at_issue': 0})
```

```python
# Replace missing values with 0 in column: 'total_recovered_on_time'
loan = loan.fillna({'total_recovered_on_time': 0})
test = test.fillna({'total_recovered_on_time': 0})
```

```python
# Replace missing values with 0 in column: 'total_recovered_15_dpd'
loan = loan.fillna({'total_recovered_15_dpd': 0})
test = test.fillna({'total_recovered_15_dpd': 0})
```

```python
# Replace missing values with 0 in column: 'cash_yield_15_dpd'
loan = loan.fillna({'cash_yield_15_dpd': 0})
test = test.fillna({'cash_yield_15_dpd': 0})
```

```python
loan = loan.fillna({'loan_number': 0})
test = test.fillna({'loan_number': 0})
```

### Drop duplicates

```python
duplicates = loan[loan.duplicated()]

# Print duplicate rows
duplicates.head(10).style
```

```python
loan = loan.drop_duplicates()
```

### Drop columns

Assume that the credit officer id and aqusition channel has any relevance of getting the loan aporoved.

```python
# Drop columns: 'credit_officer_id', 'acquisition_channel','employee_count'
loan = loan.drop(columns=['credit_officer_id', 'acquisition_channel','employee_count'])

# Drop columns: 'credit_officer_id', 'acquisition_channel', 'payment_status','employee_count'
test = test.drop(columns=['credit_officer_id', 'acquisition_channel', 'employee_count', 'payment_status'])
```

### Change categorical feature to numerical.

- Find categorical features
- Find unique values for each feature
- Match each unique value to a range of 1 to the length of the list of unique values.
- Apply ordinal encoding to the each categorical feature.

```python
categorical_columns = loan.select_dtypes(include=['object', 'category']).columns
categorical_columns
```

```python
unique_dismissal_values = loan['dismissal_description'].unique()
unique_dismissal_values
```

```python
unique_dismissal_dict = {value: index + 1 for index, value in enumerate(unique_dismissal_values)}

# Display the dictionary
print(unique_dismissal_dict)
```

```python
dismissal_encoder = OrdinalEncoder(cols=['dismissal_description'], 
                            mapping=[{'col': 'dismissal_description', 'mapping': unique_dismissal_dict}])

loan = dismissal_encoder.fit_transform(loan)
test = dismissal_encoder.fit_transform(test)
```

```python
unique_sector_values = loan['sector'].unique()
unique_sector_values
```

```python
unique_sector_dict = {value: index + 1 for index, value in enumerate(unique_sector_values)}

# Display the dictionary
print(unique_sector_dict)
```

```python
sector_encoder = OrdinalEncoder(cols=['sector'], 
                            mapping=[{'col': 'sector', 'mapping': unique_sector_dict}])

loan = sector_encoder.fit_transform(loan)
test = sector_encoder.fit_transform(test)
```

```python
unique_approval_values = loan['approval_status'].unique()
unique_approval_values
```

```python
approval_encoder = OrdinalEncoder(cols=['approval_status'], 
                            mapping=[{'col': 'approval_status', 'mapping': {'Approved': 1, 'Declined': 2, 'Cancelled': 3, 'Expired': 4}}])

loan = approval_encoder.fit_transform(loan)
test = approval_encoder.fit_transform(test)
```

### Change binary values to numerical

```python
# Replace all instances of False with 0 in column: 'paid_late'
# Replace all instances of True with 1 in column: 'paid_late'
loan['paid_late'] = loan['paid_late'].replace({False: 0, True: 1})
test['paid_late'] = test['paid_late'].replace({False: 0, True: 1})
```
